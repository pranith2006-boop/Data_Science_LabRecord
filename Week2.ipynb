```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
df
```

```python
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35]
})
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
df = pd.DataFrame({
    'Sales': [100, np.nan, np.nan, 250, 300]
})
df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)
df_ffill
```

```python
df = pd.DataFrame({
    'Sales': [100, np.nan, np.nan, 250, 300]
})
df_bfill = df.copy()
df_bfill.fillna(method='bfill', inplace=True)
df_bfill
```

```python
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})
df_drop_rows = df.dropna()
df_drop_rows
```

```python
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})
df_drop_cols = df.dropna(axis=1)
df_drop_cols
```

```python
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Salary': [50000, np.nan, 60000, 70000, np.nan],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})
df
df_mean = df.copy()
df_mean['Age'].fillna(df_mean['Age'].mean(), inplace=True)
df_median = df.copy()
df_median['Salary'].fillna(df_median['Salary'].median(), inplace=True)
df_mode = df.copy()
df_mode['Department'].fillna(df_mode['Department'].mode()[0], inplace=True)
df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)
df_bfill = df.copy()
df_bfill.fillna(method='bfill', inplace=True)
df_drop_rows = df.dropna()
df_drop_cols = df.dropna(axis=1)
df_mean
df_median
df_mode
df_ffill
df_bfill
df_drop_rows
df_drop_cols
```
