```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ztest
from scipy.stats import spearmanr
from scipy import stats
from scipy.stats import chi2_contingency
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
mean_values = df.mean()
mean_values
median_values = df.median()
median_values
mode_values = df.mode().iloc[0]
mode_values
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
range_values = df.max() - df.min()
range_values
variance_values = df.var()
variance_values
std_values = df.std()
std_values
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
skewness_values = df.skew()
print(skewness_values)
kurtosis_values = df.kurt()
kurtosis_values
```

```python
iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
corr_matrix = df.corr(method='pearson')
corr_matrix
```

```python
iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
corr_matrix = df.corr(method='spearman')
corr_matrix
```

```python
data = pd.DataFrame({
    'Gender':['M','F','M','F','M'],
    'Purchase':['Yes','No','Yes','No','Yes']
})
table = pd.crosstab(
    data['Gender'],
    data['Purchase']
)
chi2, p, dof, expected = chi2_contingency(table)
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
```

```python
group1 = [12,15,14,10,13]
group2 = [22,25,24,20,23]
t_stat, p_value = ttest_ind(
    group1,
    group2
)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
group1 = [10,12,11]
group2 = [20,22,19]
group3 = [30,31,29]
f_stat, p_value = f_oneway(
    group1,
    group2,
    group3
)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
sample = np.array(
    [10,12,11,13,12,14,15]
)
z_stat, p_value = ztest(
    sample,
    value=12
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
data = [10,12,11,13,14,100]
z_scores = np.abs(
    stats.zscore(data)
)
outliers = np.where(
    z_scores > 3
)
print("Outlier Index:", outliers)
```

```python
df = pd.DataFrame({
    'Values':[10,12,11,13,14,100]
})

Q1 = df['Values'].quantile(0.25)
Q3 = df['Values'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[
    (df['Values'] < lower) |
    (df['Values'] > upper)
]
print("Outliers:")
print(outliers)
```

```python
data = np.array([
    10,12,11,13,14,100
])
mean = np.mean(data)
std = np.std(data)
G = np.max(
    np.abs(data - mean)
) / std
print("Grubbs Statistic:", G)
```
