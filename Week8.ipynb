```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)
plt.figure(figsize=(6,4))
sns.histplot(
    df['sepal length (cm)'],
    bins=20,
    kde=True
)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x=df['sepal length (cm)']
)

plt.title("Boxplot of Sepal Length")

plt.show()
```

```python
df['species'] = iris.target
species_counts = df['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    species_counts,
    labels=iris.target_names,
    autopct='%1.1f%%'
)
plt.title(
    "Species Distribution"
)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x=df['sepal length (cm)'],
    y=df['sepal width (cm)'],
    hue=df['species']
)
plt.title(
    "Sepal Length vs Sepal Width"
)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(
    df['sepal length (cm)']
)
plt.title(
    "Line Chart of Sepal Length"
)
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(
    x=df['species'],
    y=df['sepal length (cm)']
)
plt.title(
    "Mean Sepal Length by Species"
)
plt.xticks(
    [0,1,2],
    iris.target_names
)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df.iloc[:,:4].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title(
    "Feature Correlation Heatmap"
)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(
    df['sepal length (cm)'],
    df['sepal width (cm)'],
    s=df['petal length (cm)'] * 20,
    alpha=0.5,
    c=df['species']
)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title(
    "Bubble Chart"
)
plt.show()
```

```python
sns.pairplot(
    df.iloc[:,:4]
)
plt.suptitle(
    "Pair Plot of Iris Features",
    y=1.02
)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(
    x=df['species'],
    y=df['sepal length (cm)']
)
plt.title(
    "Violin Plot: Sepal Length by Species"
)
plt.xticks(
    [0,1,2],
    iris.target_names
)
plt.show()
```
